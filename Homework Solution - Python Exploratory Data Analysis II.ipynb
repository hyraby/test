## Import Library

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load dataset

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df = df[df.notnull()]
df['ChurnNumeric'] = np.where(df['Churn']=='Yes',1,0)
df = df[df[['TotalCharges']].apply(lambda x: x[0].isdigit(), axis=1)]
df['TotalCharges'] = df['TotalCharges'].astype(float)
df.head()
```

## Soal

### Soal 1

Assessment criteria - Tampilkan data customerID dengan filter :<br><br>
nilai TotalCharges-nya di atas nilai mean/average<br>
nilai MonthlyCharges-nya di atas nilai percentile 60<br>
PaymentMethod hanya berasal dari “Electronic check”<br>
customerID tidak di awali oleh “0”<br>

```python
mean_TotalCharges = df['TotalCharges'].mean()
P60_MonthlyCharges =  df['MonthlyCharges'].quantile(0.6)
mean_TotalCharges, P60_MonthlyCharges
```

```python
##Hasil harus seperti ini, terdiri dari 26 rows dan 22 columns

df[(df['TotalCharges']>=mean_TotalCharges)
   & (df['MonthlyCharges']>=P60_MonthlyCharges)
   & (df['PaymentMethod'].isin(['Electronic check']))
   & (~df['customerID'].str.startswith('0'))
]
```

```python
df[(df['PaymentMethod'] == 'Electronic check')
   &(~df['customerID'].str.startswith('0'))
]
```

### Soal 2

Assessment criteria - Lakukan aggregate data (Group-by aggregation atau Cross-tabulation)dengan kolom<br><br>
PaymentMethod dan InternetService kemudian lihat nilai mean dari MonthlyCharges<br>




```python
grouped = df.groupby(['PaymentMethod', 'InternetService'])['MonthlyCharges'].mean().round(0)
grouped
```
